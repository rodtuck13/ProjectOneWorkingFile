```python
from dotenv import load_dotenv
import os
import requests
import json
import pandas as pd
import quandl
```

```python
sf_neighborhood_codes = pd.read_csv('san_francisco_neighborhood_codes.csv')
sf_neighborhood_codes.head()
```

```python
num_of_regions = len(sf_neighborhood_codes.index)
num_of_regions
```

```python
load_dotenv()
```

```python
quandl.ApiConfig.api_key = os.getenv('QUANDL_API_KEY')
```

```python
data = {}
count = 0
while count < num_of_regions:
    neighbor_code = sf_neighborhood_codes.iloc[count,0]
    data[count] = quandl.get_table('ZILLOW/DATA', indicator_id='Z1BR', region_id=neighbor_code)
    data[count] = data[count].set_index('date')
    count += 1
```

```python
data[3].head()
```

```python
first_data = 'yes'
count1 = 0
dataframe_count = 0
sf_1br = pd.DataFrame()

while count1 < num_of_regions:
    if len(data[count1].index) == 0 and first_data == 'yes':
        count1 += 1
    elif len(data[count1].index) > 0 and first_data == 'yes':
        dataframe_count = 1
        #sf_1br['date'] = data[count1]['date']
        sf_1br['indicator_id'] = data[count1]['indicator_id']
        sf_1br['value'] = data[count1]['value']
        sf_1br['count'] = dataframe_count
        first_data = 'no'
        count1 += 1
    elif len(data[count1].index) == 0 and first_data == 'no':
        count1 += 1
    elif len(data[count1].index) > 0 and first_data == 'no':
        dataframe_count += 1
        sf_1br['value'] += data[count1]['value']
        sf_1br['count'] = dataframe_count
        count1 += 1

sf_1br['mean'] = sf_1br['value']/sf_1br['count']
sf_1br
```

```python
sf_1br.to_csv(r'san_francisco_1br.csv')
```
